```python
import pandas as pd
```

```python
## load pass data
pass_data = pd.read_csv("nfl_passing_stats.csv")
pass_df = pd.DataFrame(pass_data)
pass_df.head()
```

```python
#drop columns
pass_df = pass_df.drop(
    ['rate', '1st', '1st_Percentage', '20plus', '40plus', 'lng', 'sck', 'scky'], axis=1)
pass_df.head()
```

```python
# rename columns
pass_df.rename(columns={"yds_att": "yds/att", "passyds": "pass_yds", "int_":"int"}, inplace=True)
pass_df.head()
```

```python
## load rush data
rush_data = pd.read_csv("nfl_rushing_stats.csv")
rush_df = pd.DataFrame(rush_data)
rush_df.head()
```

```python
## drop columns
rush_df = rush_df.drop(
    ['20plus', '40plus', 'lng', 'rush_1st', 'rush_1st_percentage'], axis=1)
rush_df.head()
```

```python
## rename columns
rush_df.rename(columns={"rushyds": "rush_yds", "rushfum": "fum"}, inplace=True)
rush_df.head()
```

```python
## merge columns
offense_df = pass_df.merge(rush_df, left_index=True, right_index=True)
offense_df.head()
```

```python
## drop and rename columns on merged df
offense_df = offense_df.drop("team_y",1)
offense_df = offense_df.drop("year_y",1)
offense_df.rename(columns={"team_x": "team", "att_x": "pass_att", "td_x": "pass_td", "year_x": "year", "att_y": "rush_att", "td_y": "rush_td"}, inplace=True)
offense_df.head()
```

```python
## drop duplicate team name
teams = offense_df.team
i = 0
for team in offense_df.team:
    teams[i] = team[:len(team)//2]
    i = i + 1
    
offense_df["team"] = teams
offense_df.head()
```

```python
#checking for washington name change.  seeing if they are called Football Team or Redskins
offense_df["team"].unique()
```

```python
#making only one name for washington
was_teams = offense_df.team
i = 0
for team in offense_df.team:
    if team == "Redskins":
        was_teams[i] = "Football Team"
    i = i + 1
    
was_teams.unique()
```

```python
# correcting df
offense_df["team"] = was_teams
offense_df.head()
```

```python
# add playoff column
offense_df["playoff"] = int
```

```python
offense_df.to_csv("cleaned_offense.csv")
```
