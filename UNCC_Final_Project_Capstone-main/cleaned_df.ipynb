```python
import pandas as pd

scraper = pd.read_html("https://www.nfl.com/stats/team-stats/offense/passing/2021/reg/all")

for idx, table in enumerate(scraper):
    print("*********************")
    print(idx)
    print(table)
```

```python
scraper[0]
```

```python
stats_df = scraper[0]
stats_df.head()
```

```python
stats_df = stats_df.drop("Cmp %", 1)
stats_df = stats_df.drop("Rate", 1)
stats_df = stats_df.drop("1st", 1)
stats_df = stats_df.drop("1st%", 1)
stats_df = stats_df.drop("20+", 1)
stats_df = stats_df.drop("40+", 1)
stats_df = stats_df.drop("Lng", 1)
stats_df = stats_df.drop("Sck", 1)
stats_df = stats_df.drop("SckY", 1)
stats_df.head()
```

```python
rush_scraper = pd.read_html("https://www.nfl.com/stats/team-stats/offense/rushing/2021/reg/all")

for idx, table in enumerate(rush_scraper):
    print("*********************")
    print(idx)
    print(table)
```

```python
rushing_df = rush_scraper[0]
rushing_df.head()
```

```python
rushing_df = rushing_df.drop("20+",1)
rushing_df = rushing_df.drop("40+",1)
rushing_df = rushing_df.drop("Lng",1)
rushing_df = rushing_df.drop("Rush 1st",1)
rushing_df = rushing_df.drop("Rush 1st%",1)
rushing_df = rushing_df.drop("Rush FUM",1)

rushing_df.head()
```

```python
offensive_stats_df = stats_df.merge(rushing_df, left_index=True, right_index=True)
offensive_stats_df.head()
```

```python
offensive_stats_df = offensive_stats_df.drop("Team_y", 1)
offensive_stats_df.head()
```

```python
# relabel columns
offensive_stats_df.rename(columns={"Team_x": "Team", "Att_x": "Pass_att", "Pass Yds": "Pass_yds", "TD_x":"Pass_TD", "Att_y":"Rush_att", "Rush Yds":"Rush_yds", "TD_y":"Rush_TD"}, inplace=True)
offensive_stats_df.head()
```

```python
# Team Name index
team_list = offensive_stats_df.Team
print(team_list)
```

```python
# drop duplicate
i = 0
for team in offensive_stats_df.Team:
    team_list[i] = team[:len(team)//2]
    i = i+1
print(team_list)
```

```python
# replace Team column with Team list
offensive_stats_df["Team"] = team_list
offensive_stats_df.head()
```
